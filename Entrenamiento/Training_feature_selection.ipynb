```python
import pandas as pd
from sklearn import svm
from sklearn.utils import shuffle
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("..\\Base de datos\\Feature_extraction_extended.csv")
```

```python
#X = data.drop("Label", axis=1)  # Ajusta la columna de etiquetas según tu DataFrame
X = data.drop (['Image ID', 'Label'], axis = 1)
y = data["Label"]
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba con el mismo barajeo
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Imprimir el tamaño de los conjuntos de entrenamiento y prueba
print("Tamaño del conjunto de entrenamiento:", X_train.shape)
print("Tamaño del conjunto de prueba:", X_test.shape)
```

```python
# Crear el modelo Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Entrenar el modelo
model.fit(X_train, y_train)

# Predecir las etiquetas para el conjunto de prueba
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Precisión del modelo: {:.2f}%".format(accuracy * 100))
```

```python
# Obtener la importancia de las características
importancia_caracteristicas = model.feature_importances_

# Crear un DataFrame con las características y su importancia
df_importancia = pd.DataFrame({'Característica': X.columns, 'Importancia': importancia_caracteristicas})

# Ordenar las características por su importancia de forma descendente
df_importancia = df_importancia.sort_values('Importancia', ascending=False)

# Visualizar las características más importantes
plt.figure(figsize=(10, 6))
plt.bar(df_importancia['Característica'], df_importancia['Importancia'])
plt.xticks(rotation=90)
plt.xlabel('Característica')
plt.ylabel('Importancia')
plt.title('Importancia de las características en el modelo Random Forest')
plt.show()
#plt.savefig("Importancia de las caracteristicas en el modelo Random Forest")
```

```python
data = pd.read_csv("..\\Base de datos\\Feature_extraction_extended_RF.csv")
#X = data.drop("Label", axis=1)  # Ajusta la columna de etiquetas según tu DataFrame
X = data.drop (['Image ID', 'Label'], axis = 1)
y = data["Label"]
# Dividir los datos en conjuntos de entrenamiento y prueba con el mismo barajeo
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Imprimir el tamaño de los conjuntos de entrenamiento y prueba
print("Tamaño del conjunto de entrenamiento:", X_train.shape)
print("Tamaño del conjunto de prueba:", X_test.shape)
```

```python
model = svm.SVC()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Precisión del modelo SVM:{:.2f}%".format(accuracy * 100))
```

```python
# Crear el modelo KNN
model = KNeighborsClassifier(n_neighbors=5)
# Entrenar el modelo
model.fit(X_train, y_train)
# Predecir las etiquetas para el conjunto de prueba
y_pred = model.predict(X_test)
# Calcular la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)
print("Precisión del modelo: {:.2f}%".format(accuracy * 100))
```
